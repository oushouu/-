```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import StandardScaler

import keras
from keras.models import Sequential
from keras.layers import Dense

from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
train = pd.read_csv('TREE_DATA_with_WBS_train.csv')
test1 = pd.read_csv("TREE_DATA_with_WBS_test1.csv") 
test1s = pd.read_csv("TREE_DATA_with_WBS_test1s.csv") 
test2 = pd.read_csv("TREE_DATA_with_WBS_test2.csv") 
test2s = pd.read_csv("TREE_DATA_with_WBS_test2s.csv") 
```

```python
X_train = train.iloc[:, 3:].drop(['ロスコン案件マーク', '月度', '原価率進捗'], axis=1).values
y_train = train.iloc[:, 5].values
```

```python
X_test1 = test1.iloc[:, 3:].drop(['ロスコン案件マーク', '月度', '原価率進捗'], axis=1).values
X_test1s = test1s.iloc[:, 3:].drop(['ロスコン案件マーク', '月度', '原価率進捗'], axis=1).values
X_test2 = test2.iloc[:, 3:].drop(['ロスコン案件マーク', '月度', '原価率進捗'], axis=1).values
X_test2s = test2s.iloc[:, 3:].drop(['ロスコン案件マーク', '月度', '原価率進捗'], axis=1).values
```

```python
# Feature Scaling 标准化
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
X_test1 = sc.transform(X_test1)
X_test1s = sc.transform(X_test1s)
X_test2 = sc.transform(X_test2)
X_test2s = sc.transform(X_test2s)
```

```python
# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 32, epochs = 500)
```

```python
# Predicting the Test set results
y_pred_test1 = classifier.predict(X_test1)
y_pred_test1s = classifier.predict(X_test1s)
y_pred_test2 = classifier.predict(X_test2)
y_pred_test2s = classifier.predict(X_test2s)
```

```python
ANN_test1_submit = pd.concat([test1, pd.DataFrame(y_pred_test1).rename({0: 'ANN_score'}, axis='columns')], axis=1).iloc[:, [1,2,-1]]
ANN_test1s_submit = pd.concat([test1s, pd.DataFrame(y_pred_test1s).rename({0: 'ANN_score'}, axis='columns')], axis=1).iloc[:, [1,2,-1]]
ANN_test2_submit = pd.concat([test2, pd.DataFrame(y_pred_test2).rename({0: 'ANN_score'}, axis='columns')], axis=1).iloc[:, [1,2,-1]]
ANN_test2s_submit = pd.concat([test2s, pd.DataFrame(y_pred_test2s).rename({0: 'ANN_score'}, axis='columns')], axis=1).iloc[:, [1,2,-1]]
```

```python
ANN_test1_submit.to_csv("ANN_test1_submit.csv", encoding = "ANSI")
ANN_test1s_submit.to_csv("ANN_test1s_submit.csv", encoding = "ANSI")
ANN_test2_submit.to_csv("ANN_test2_submit.csv", encoding = "ANSI")
ANN_test2s_submit.to_csv("ANN_test2s_submit.csv", encoding = "ANSI")
```
