```python
import pandas as pd
```

```python
devdata_file = pd.ExcelFile("devdata.xlsx")
```

```python
devdata_file
```

```python
devdata = pd.read_excel(devdata_file)
```

```python
devdata.tail()
```

```python
# replace nan with '' in ACCOUNT_NAME
devdata.ACCOUNT_NAME.fillna('', inplace=True)
```

```python
# put all characters into a big_pool\
big_pool = []
for i in range(len(devdata)):
    big_pool= big_pool + list(devdata["ACCOUNT_NAME"][i])
```

```python
# extract unique elements from big pool
distinct_pool = set(big_pool)
```

```python
len(distinct_pool)
```

```python
# extract unique group name from ACCOUNT_GROUP
group_name_set = set(list(devdata["ACCOUNT_GROUP"]))
```

```python
# for each group, create group pools
group_pool_dict = dict()
for group_name in group_name_set:
    inner_group_pool = list()
    for i in range(len(devdata)): 
        if devdata["ACCOUNT_GROUP"][i] == group_name:
            inner_group_pool= inner_group_pool + list(devdata["ACCOUNT_NAME"][i])
    group_pool_dict[group_name] = inner_group_pool
            
```

```python
# example
len(group_pool_dict["営業外費用"])
```

```python
# create frequency matrix of i char in j group
char_freq_dict = dict()
for char in distinct_pool:
    char_freq_dict[char] = []
    for key in group_pool_dict:
        char_freq_dict[char].append(group_pool_dict[key].count(char)/len(group_pool_dict[key]))
```

```python
# convert dict to df
char_freq_df = pd.DataFrame.from_dict(char_freq_dict)
```

```python
# add ACCOUNT_GROUP column as join key
char_freq_df["ACCOUNT_GROUP"] = group_name_set
```
