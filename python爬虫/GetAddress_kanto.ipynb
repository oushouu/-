```python
import requests, bs4, webbrowser
import pandas as pd
```

```python
url = "https://www.itoyokado.co.jp/store/result.html#area_kanto"
page1 = requests.get(url)
bspage1 = bs4.BeautifulSoup(page1.content, 'html.parser')
```

```python
# 関東地方7地域のdivタッグ取得
div_tokyo = bspage1.find("div", { "data-area-target":'["tokyo","area_kanto"]'})
div_kanagawa = bspage1.find("div", { "data-area-target":'["kanagawa","area_kanto"]'})
div_chiba = bspage1.find("div", { "data-area-target":'["chiba","area_kanto"]'})
div_saitama = bspage1.find("div", { "data-area-target":'["saitama","area_kanto"]'})
div_ibaraki = bspage1.find("div", { "data-area-target":'["ibaraki","area_kanto"]'})
div_tochigi = bspage1.find("div", { "data-area-target":'["tochigi","area_kanto"]'})
div_gunma = bspage1.find("div", { "data-area-target":'["gunma","area_kanto"]'})
```

```python
def GetAddress(div_area):
    AddressDict = dict()
    div_area_shop_list = div_area.find_all("div", {"class" : "-c1"})
    for shop_info in div_area_shop_list:
        shop_name = shop_info.find("a", {"class": "-arrow"}).get_text(strip=True)
        try:
            shop_address = shop_info.find("p", {"class": "-txt"}).get_text(strip=True)
        except AttributeError:
            shop_address = "No Address"
            item = {shop_name: shop_address}
        else:
            item = {shop_name: shop_address}
        AddressDict.update(item)
    return AddressDict
```

```python
# Example
GetAddress(div_tokyo)
```

```python
dict_tokyo = GetAddress(div_tokyo)
dict_kanagawa = GetAddress(div_kanagawa)
dict_chiba = GetAddress(div_chiba)
dict_saitama = GetAddress(div_saitama)
dict_ibaraki = GetAddress(div_ibaraki)
dict_tochigi = GetAddress(div_tochigi)
dict_gunma = GetAddress(div_gunma)
```

```python
dict_kanto = {**dict_tokyo, **dict_kanagawa, **dict_chiba, **dict_saitama, **dict_ibaraki, **dict_tochigi, **dict_gunma}
```

```python
dict_kanto
```

```python
df_kanto = pd.DataFrame.from_dict(dict_kanto, orient='index',
                       columns=['post_address'])
```

```python
df_kanto
```

```python
df_kanto.to_csv("df_kanto.csv", encoding = "ANSI")
```

```python
# Rで整理
PostAddress_kanto <- read.csv("df_kanto.csv") %>% 
  filter(post_address != "No Address") %>% 
  mutate(PostCode = str_sub(post_address, 1,9), 
         Address = str_sub(post_address, 11, -1)) %>% 
  select(-post_address)
  
write.table(PostAddress_kanto, file='PostAddress_kanto.tsv', quote=FALSE, sep='\t', col.names = NA)
```
