```python
# web crawler用
import requests, bs4, webbrowser

# 正規表現用
import re 
```

```python
# Google Mapsで、"ito-yokado 赤羽店"というキーボードを検索結果ページのHTML内容を取る
url = "https://www.google.com/maps/search/" + "イトーヨーカドー 田無店"
page1 = requests.get(url)
bspage1 = bs4.BeautifulSoup(page1.content, 'html.parser')
```

```python
# 検索結果ページのHTML内容一覧
bspage1
```

```python
# HTML内容から目標の対象タッグを取る
target = bspage1.find_all('meta', {'content': re.compile('.*center.*')})[0]
target
```

```python
# 目標の対象タッグから経緯度が含むurlのストリングを取得
target.get('content')
```

```python
# 一つ目の経緯度を取る
lat_lon =re.findall("(?<=center=).*(?=\&zoom)" , target.get('content'))
lat_lon
```

```python
lat = re.findall("^.*(?=%2C)" , lat_lon[0])[0]
lon = re.findall("(?<=%2C).*$" , lat_lon[0])[0]
```
