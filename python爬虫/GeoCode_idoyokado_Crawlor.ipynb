```python
# web crawler用
import requests, bs4, webbrowser

# 正規表現用
import re 

import pandas as pd
```

```python
selected_shop_name = pd.read_csv("selected_shop_name.csv")
```

```python
selected_shop_name
```

```python
keyword_list = list(selected_shop_name["FullName"])
```

```python
# キーボード検索結果の経緯度を取得関数を定義する
def getGoogleGeoCode(key_word):
    
    url = "https://www.google.com/maps/search/" + key_word
    page1 = requests.get(url)
    bspage1 = bs4.BeautifulSoup(page1.content, 'html.parser')
    # 検索結果存在するかどうかを試し
    try:
        target = bspage1.find_all('meta', {'content': re.compile('.*center.*')})[0]
    # 検索結果存在しないエラー発生の処理   
    except IndexError:
        print(key_word)
        return {key_word: [0, 0]}
    # エラー無かったら
    else:
        lat_lon =re.findall("(?<=center=).*(?=\&zoom)" , target.get('content'))
        lat = re.findall("^.*(?=%2C)" , lat_lon[0])[0]
        lon = re.findall("(?<=%2C).*$" , lat_lon[0])[0]
        return {key_word: [lat, lon]}
```

```python
# エラー発生の例
getGoogleGeoCode("イトーヨーカドー ＳＳ秋田店")
```

```python
keyword_list
```

```python
GeoDict = dict()
index = 0
for keyword in keyword_list:
    print(index)
    GeoDict.update(getGoogleGeoCode(keyword))
    index+= 1       
```

```python
GeoDict
```

```python
GeoCode_df = pd.DataFrame(list(GeoDict.items()), columns = ['FullName','GeoCode'])
```

```python
GeoCode_df
```

```python
GeoCode_df[['lat','lon']] = pd.DataFrame(GeoCode_df.GeoCode.tolist(), index= GeoCode_df.index)
```

```python
GeoCode_df.to_csv("GeoCode_df.csv", encoding = "ANSI")
```
